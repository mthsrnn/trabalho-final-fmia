```python
#imports básicos
import pandas as pd
import numpy as np

#pre-processamento (via scikit)
from sklearn.model_selection import train_test_split

#mlp classifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

#métricas e plots
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('arcos_gerados.csv', converters={'entrada': str})
data.head()
```

```python
X, y = data['entrada'], data['saida']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

X_train = X_train.str.split('', expand=True).iloc[:, 1:-1]
X_test = X_test.str.split('', expand=True).iloc[:, 1:-1]
y_train = y_train.str.split('', expand=True).iloc[:, 1:-1]
y_test = y_test.str.split('', expand=True).iloc[:, 1:-1]

X_train = X_train.astype(float)
X_test = X_test.astype(float)
y_train = y_train.astype(float)
y_test = y_test.astype(float)

X_train.head()
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.metrics import make_scorer, mean_squared_error, r2_score

classifier = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=100, activation='relu', solver='adam', random_state=1)

# Define the scoring metrics
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
r2_scorer = make_scorer(r2_score)

# Set up cross-validation with KFold
kf = KFold(n_splits=5, shuffle=True, random_state=1)  # 5-fold cross-validation

# Perform cross-validation
mse_scores = cross_val_score(classifier, X_train, y_train, cv=kf, scoring=mse_scorer)
r2_scores = cross_val_score(classifier, X_train, y_train, cv=kf, scoring=r2_scorer)

# Train and predict using cross-validation
y_pred_cv = cross_val_predict(classifier, X_train, y_train, cv=kf)

# Calculate mean squared error and R-squared for the predictions
mse_cv = mean_squared_error(y_train, y_pred_cv)
r2_cv = r2_score(y_train, y_pred_cv)
accuracy = accuracy_score(y_train, y_pred_cv)

print("Cross-Validated Mean Squared Error:", -mse_scores.mean())  # Negate because scoring is negated MSE
print("Cross-Validated R-squared:", r2_scores.mean())
print("Overall Mean Squared Error from CV predictions:", mse_cv)
print("Overall R-squared from CV predictions:", r2_cv)
```
